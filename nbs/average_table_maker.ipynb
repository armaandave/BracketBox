```python
#imports
import pandas as pd 
from tqdm.notebook import tqdm
```

```python
def get_yearly_avgs():
    df = pd.read_csv('/Users/armaan/src/bbox/mdata/final_result/master_table.csv') 
    
    years_avgs = []
    for yr in df.year.unique():
        yr_avg = {'year' : yr}
        yr_avg['bb_score'] = df[df.year == yr].bb_score.mean()
        yr_avg['ww_box_office'] = df[df.year == yr].worldwide_box_office.mean()
        yr_avg['production_budget'] = df[df.year == yr].production_budget.mean()
        yr_avg['bb_profit'] = df[df.year == yr].bb_profit.mean()

        years_avgs.append(yr_avg)

    return pd.DataFrame(years_avgs)
    
dfyears = get_yearly_avgs()
dfyears    
```

```python
dfyears.to_csv('/Users/armaan/src/bbox/mdata/final_result/yearly_avgs.csv', index=False)
```
